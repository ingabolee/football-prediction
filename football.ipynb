# A machine learning project that uses deep neural networks for football prediction

**Importing the required libraries**

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import numpy as np
from google.colab import files
```

**Uploading the data file**

```python
uploaded = files.upload()
df = pd.read_csv('finals.csv')
```

Dropping of the columns that are to be predicted and,
specifying the output column

```python
properties = list(df.columns.values)
properties.remove('o35')

X = df[properties]
y = df['o35']
```

Splitting the dataset into training and testing categories and, describing the shape of the neural network.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=6)

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(19,)),
    keras.layers.Dense(64, activation=tf.nn.relu),
    keras.layers.Dense(64, activation=tf.nn.tanh),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid),
])
```

Compiling the model by specifyng the optimizer and loss functions. Training of the model

```python

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5000, batch_size=1)
```

Testing the model and printing its accuracy score.

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Test accuracy: {test_acc*100}%')
```
